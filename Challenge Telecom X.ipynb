```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import requests
import matplotlib.ticker as ticker

# Cargar datos
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()

# Normalización
df = pd.DataFrame(data)
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])
df = pd.concat([df, df_customer, df_phone, df_internet, df_account], axis=1)
df = df.drop(['customer', 'phone', 'internet', 'account'], axis=1)

# Limpieza
df['Churn'] = df['Churn'].replace('', 'No')
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
df['Charges.Monthly'] = pd.to_numeric(df['Charges.Monthly'], errors='coerce')
df['Charges.Total'] = df['Charges.Total'].fillna(df['Charges.Total'].mean())
df['Cuentas_Diarias'] = df['Charges.Monthly'] / 30

# Configuración visual
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 13

# Diccionario de títulos descriptivos y formato para ejes
column_info = {
    'tenure': ('Antigüedad del Cliente', 'Meses'),
    'Charges.Monthly': ('Cargos Mensuales', '$ Pesos Colombianos'),
    'Charges.Total': ('Cargos Totales', '$ Pesos Colombianos'),
    'Cuentas_Diarias': ('Cobro Diario Promedio', '$ Pesos por Día')
}

# Colores fijos para consistencia
palette = {'No': '#2ca02c', 'Yes': '#d62728'}

for col, (titulo, unidad) in column_info.items():
    if col not in df.columns:
        continue

    # KDE
    plt.figure(figsize=(10, 6))
    ax = sns.kdeplot(data=df, x=col, hue='Churn', fill=True, palette=palette, common_norm=False, alpha=0.4, linewidth=2)
    plt.title(f'Distribución de {titulo} ({unidad}) según Evasión de Clientes')
    plt.xlabel(f'{titulo} ({unidad})')
    plt.ylabel('Densidad de Clientes')
    plt.legend(title='¿Cliente Canceló Servicio?', labels=['No Evadió', 'Evadió'])
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
    plt.tight_layout()
    plt.show()

    # BOX PLOT
    plt.figure(figsize=(10, 6))
    ax = sns.boxplot(data=df, x='Churn', y=col, palette=palette, width=0.5)
    plt.title(f'{titulo} por Estado de Evasión')
    plt.xlabel('Estado de Evasión del Cliente')
    plt.ylabel(f'{titulo} ({unidad})')
    plt.xticks(ticks=[0, 1], labels=['No Evadió', 'Evadió'])
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
    plt.tight_layout()
    plt.show()
```

**Informe de Análisis de Evasión de Clientes (Churn) - Proyecto Telecom X**


Aquí les presento un resumen de todo el trabajo que realizamos en este proyecto de análisis de evasión de clientes. El objetivo ha sido claro desde el inicio: entender por qué algunos de nuestros clientes se van, y qué podemos hacer para que eso ocurra menos.

🔹 **Introducción**
La pérdida de clientes, también conocida como "Churn", es uno de los retos más grandes para cualquier empresa de servicios. Cada cliente que se va representa ingresos que no volverán, y reemplazarlo cuesta tiempo, dinero y esfuerzo. Por eso es tan importante identificar patrones que nos permitan anticiparnos.

Este análisis se enfocó en detectar qué factores influyen más en la decisión de cancelar el servicio, usando datos reales de nuestra base de clientes. El propósito no era solo observar números, sino sacar conclusiones útiles que puedan convertirse en acciones concretas.

🔹 **Limpieza y Tratamiento de Datos**
Antes de poder analizar cualquier cosa, tuvimos que poner orden en los datos. Fue un proceso importante que nos aseguró trabajar sobre una base confiable. Estos fueron los pasos principales:

Importación: Trajimos los datos desde un archivo JSON.

Estructuración: El archivo contenía información anidada por secciones (cliente, teléfono, internet, cuenta). Desarmamos esa estructura para tener todo en un formato plano y fácil de leer.

Depuración: Quitamos columnas duplicadas o innecesarias tras la normalización.

Corrección de errores en ‘Churn’: Detectamos algunos valores vacíos en la columna objetivo. Suponiendo que representaban clientes activos, los etiquetamos como “No”.

Conversión de datos numéricos: La columna de cargos totales (‘Charges.Total’) tenía valores no válidos. Convertimos todo a números y rellenamos los valores faltantes con el promedio.

Nueva columna - ‘Cuentas_Diarias’: Para tener una vista más granular, creamos esta columna dividiendo los cargos mensuales entre 30.

Con esto, dejamos todo listo para el análisis exploratorio.

🔹 **Análisis Exploratorio de Datos (EDA)**
Una vez que tuvimos los datos listos, nos sumergimos en explorar relaciones, diferencias y patrones. Aquí van los puntos más relevantes:

1. ¿Cuántos clientes se van?
Empezamos por lo más básico: ver la proporción de clientes que evadieron frente a los que permanecen.

📊 [Gráfico de barras de Churn]
🔍 Resultado: Solo el 26% de los clientes se fueron. Aunque parezca poco, es un número que puede crecer si no se actúa a tiempo.

2. ¿Quiénes se van más? (Variables categóricas)
Exploramos cómo se comporta la evasión según diferentes características:

📊 [Gráficos por género, tipo de contrato y método de pago]

Género: No hubo mucha diferencia entre hombres y mujeres.

Tipo de contrato: El contrato “mes a mes” es el más vulnerable. Aquí se concentra la mayoría de las cancelaciones.

Método de pago: Los clientes que pagan con cheque electrónico son los que más evaden. Esto podría tener que ver con la experiencia del pago.

3. ¿Qué más influye? (Variables numéricas)
Vimos cómo la evasión se relaciona con variables como la antigüedad del cliente, cargos mensuales y totales.

📊 [Gráficos de densidad y boxplots para tenure, cargos y cuentas diarias]

Antigüedad (tenure): Los clientes nuevos se van más. La mayoría se da de baja en los primeros meses.

Cargos mensuales: A mayor cargo, mayor tasa de evasión. Tal vez no perciben suficiente valor.

Cargos totales: Los que han pagado poco suelen ser los que se van rápido, probablemente porque son nuevos o no encontraron lo que buscaban.

🔹 **Conclusiones e Insights**
Después de revisar todos estos datos, las conclusiones más importantes fueron:

Clientes nuevos = más riesgo: Hay que cuidar especialmente a quienes recién comienzan.

Los contratos mes a mes son débiles: Dan más libertad, pero eso también facilita la salida.

El método de pago importa: Especialmente cuando es por cheque electrónico, hay que entender qué lo vuelve incómodo.

Altos cargos mensuales generan fricción: Podría ser por falta de claridad, percepción de precio elevado o mala experiencia de servicio.

En resumen: la evasión tiene patrones claros, y eso es una gran ventaja porque nos permite anticiparnos.

🔹 **Recomendaciones**
Con base en todo lo anterior, proponemos estas acciones concretas:

Fortalecer el onboarding de nuevos clientes: Los primeros 3 meses son clave. Un acompañamiento más cercano, comunicaciones claras y beneficios iniciales pueden hacer la diferencia.

Fomentar contratos de largo plazo: Ofrecer incentivos, descuentos o beneficios exclusivos para quienes elijan contratos anuales o bianuales.

Revisar la experiencia de pago con cheques electrónicos: ¿Es complicada? ¿Lenta? Una encuesta o entrevista podría revelar más.

Ofrecer opciones personalizadas a quienes pagan más: No basta con cobrar. Hay que asegurar que esos clientes sientan que su inversión vale la pena.

Crear alertas tempranas: Podemos usar estos factores para construir un sistema de monitoreo que detecte clientes con alto riesgo de evasión antes de que se vayan.
