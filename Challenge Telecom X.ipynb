```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import requests
import matplotlib.ticker as ticker

# Cargar datos
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()

# Normalizaci√≥n
df = pd.DataFrame(data)
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])
df = pd.concat([df, df_customer, df_phone, df_internet, df_account], axis=1)
df = df.drop(['customer', 'phone', 'internet', 'account'], axis=1)

# Limpieza
df['Churn'] = df['Churn'].replace('', 'No')
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
df['Charges.Monthly'] = pd.to_numeric(df['Charges.Monthly'], errors='coerce')
df['Charges.Total'] = df['Charges.Total'].fillna(df['Charges.Total'].mean())
df['Cuentas_Diarias'] = df['Charges.Monthly'] / 30

# Configuraci√≥n visual
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 13

# Diccionario de t√≠tulos descriptivos y formato para ejes
column_info = {
    'tenure': ('Antig√ºedad del Cliente', 'Meses'),
    'Charges.Monthly': ('Cargos Mensuales', '$ Pesos Colombianos'),
    'Charges.Total': ('Cargos Totales', '$ Pesos Colombianos'),
    'Cuentas_Diarias': ('Cobro Diario Promedio', '$ Pesos por D√≠a')
}

# Colores fijos para consistencia
palette = {'No': '#2ca02c', 'Yes': '#d62728'}

for col, (titulo, unidad) in column_info.items():
    if col not in df.columns:
        continue

    # KDE
    plt.figure(figsize=(10, 6))
    ax = sns.kdeplot(data=df, x=col, hue='Churn', fill=True, palette=palette, common_norm=False, alpha=0.4, linewidth=2)
    plt.title(f'Distribuci√≥n de {titulo} ({unidad}) seg√∫n Evasi√≥n de Clientes')
    plt.xlabel(f'{titulo} ({unidad})')
    plt.ylabel('Densidad de Clientes')
    plt.legend(title='¬øCliente Cancel√≥ Servicio?', labels=['No Evadi√≥', 'Evadi√≥'])
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
    plt.tight_layout()
    plt.show()

    # BOX PLOT
    plt.figure(figsize=(10, 6))
    ax = sns.boxplot(data=df, x='Churn', y=col, palette=palette, width=0.5)
    plt.title(f'{titulo} por Estado de Evasi√≥n')
    plt.xlabel('Estado de Evasi√≥n del Cliente')
    plt.ylabel(f'{titulo} ({unidad})')
    plt.xticks(ticks=[0, 1], labels=['No Evadi√≥', 'Evadi√≥'])
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
    plt.tight_layout()
    plt.show()
```

**Informe de An√°lisis de Evasi√≥n de Clientes (Churn) - Proyecto Telecom X**


Aqu√≠ les presento un resumen de todo el trabajo que realizamos en este proyecto de an√°lisis de evasi√≥n de clientes. El objetivo ha sido claro desde el inicio: entender por qu√© algunos de nuestros clientes se van, y qu√© podemos hacer para que eso ocurra menos.

üîπ **Introducci√≥n**
La p√©rdida de clientes, tambi√©n conocida como "Churn", es uno de los retos m√°s grandes para cualquier empresa de servicios. Cada cliente que se va representa ingresos que no volver√°n, y reemplazarlo cuesta tiempo, dinero y esfuerzo. Por eso es tan importante identificar patrones que nos permitan anticiparnos.

Este an√°lisis se enfoc√≥ en detectar qu√© factores influyen m√°s en la decisi√≥n de cancelar el servicio, usando datos reales de nuestra base de clientes. El prop√≥sito no era solo observar n√∫meros, sino sacar conclusiones √∫tiles que puedan convertirse en acciones concretas.

üîπ **Limpieza y Tratamiento de Datos**
Antes de poder analizar cualquier cosa, tuvimos que poner orden en los datos. Fue un proceso importante que nos asegur√≥ trabajar sobre una base confiable. Estos fueron los pasos principales:

Importaci√≥n: Trajimos los datos desde un archivo JSON.

Estructuraci√≥n: El archivo conten√≠a informaci√≥n anidada por secciones (cliente, tel√©fono, internet, cuenta). Desarmamos esa estructura para tener todo en un formato plano y f√°cil de leer.

Depuraci√≥n: Quitamos columnas duplicadas o innecesarias tras la normalizaci√≥n.

Correcci√≥n de errores en ‚ÄòChurn‚Äô: Detectamos algunos valores vac√≠os en la columna objetivo. Suponiendo que representaban clientes activos, los etiquetamos como ‚ÄúNo‚Äù.

Conversi√≥n de datos num√©ricos: La columna de cargos totales (‚ÄòCharges.Total‚Äô) ten√≠a valores no v√°lidos. Convertimos todo a n√∫meros y rellenamos los valores faltantes con el promedio.

Nueva columna - ‚ÄòCuentas_Diarias‚Äô: Para tener una vista m√°s granular, creamos esta columna dividiendo los cargos mensuales entre 30.

Con esto, dejamos todo listo para el an√°lisis exploratorio.

üîπ **An√°lisis Exploratorio de Datos (EDA)**
Una vez que tuvimos los datos listos, nos sumergimos en explorar relaciones, diferencias y patrones. Aqu√≠ van los puntos m√°s relevantes:

1. ¬øCu√°ntos clientes se van?
Empezamos por lo m√°s b√°sico: ver la proporci√≥n de clientes que evadieron frente a los que permanecen.

üìä [Gr√°fico de barras de Churn]
üîç Resultado: Solo el 26% de los clientes se fueron. Aunque parezca poco, es un n√∫mero que puede crecer si no se act√∫a a tiempo.

2. ¬øQui√©nes se van m√°s? (Variables categ√≥ricas)
Exploramos c√≥mo se comporta la evasi√≥n seg√∫n diferentes caracter√≠sticas:

üìä [Gr√°ficos por g√©nero, tipo de contrato y m√©todo de pago]

G√©nero: No hubo mucha diferencia entre hombres y mujeres.

Tipo de contrato: El contrato ‚Äúmes a mes‚Äù es el m√°s vulnerable. Aqu√≠ se concentra la mayor√≠a de las cancelaciones.

M√©todo de pago: Los clientes que pagan con cheque electr√≥nico son los que m√°s evaden. Esto podr√≠a tener que ver con la experiencia del pago.

3. ¬øQu√© m√°s influye? (Variables num√©ricas)
Vimos c√≥mo la evasi√≥n se relaciona con variables como la antig√ºedad del cliente, cargos mensuales y totales.

üìä [Gr√°ficos de densidad y boxplots para tenure, cargos y cuentas diarias]

Antig√ºedad (tenure): Los clientes nuevos se van m√°s. La mayor√≠a se da de baja en los primeros meses.

Cargos mensuales: A mayor cargo, mayor tasa de evasi√≥n. Tal vez no perciben suficiente valor.

Cargos totales: Los que han pagado poco suelen ser los que se van r√°pido, probablemente porque son nuevos o no encontraron lo que buscaban.

üîπ **Conclusiones e Insights**
Despu√©s de revisar todos estos datos, las conclusiones m√°s importantes fueron:

Clientes nuevos = m√°s riesgo: Hay que cuidar especialmente a quienes reci√©n comienzan.

Los contratos mes a mes son d√©biles: Dan m√°s libertad, pero eso tambi√©n facilita la salida.

El m√©todo de pago importa: Especialmente cuando es por cheque electr√≥nico, hay que entender qu√© lo vuelve inc√≥modo.

Altos cargos mensuales generan fricci√≥n: Podr√≠a ser por falta de claridad, percepci√≥n de precio elevado o mala experiencia de servicio.

En resumen: la evasi√≥n tiene patrones claros, y eso es una gran ventaja porque nos permite anticiparnos.

Con base en todo lo anterior, proponemos estas acciones concretas:

Crear alertas tempranas: Podemos usar estos factores para construir un sistema de monitoreo que detecte clientes con alto riesgo de evasi√≥n antes de que se vayan.
